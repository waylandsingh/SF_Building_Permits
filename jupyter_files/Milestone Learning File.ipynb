```python
#The idea here is to learn & use the requests library to get information about a crypto price

#requests library
import requests
#simplejson to parse
import simplejson as json
import pandas as pd
import numpy as np
```

```python
#API https://www.alphavantage.co/documentation/#
name = ''
payload = {
           'function' : 'DIGITAL_CURRENCY_DAILY',
           'symbol' : 'BTC',
           'market' : 'USD',
           'apikey' : '8H6F3PUCMQRPWQTX'
           }

r = requests.get('https://www.alphavantage.co/query', params=payload)
print(r.url)
print((r.text))
```

```python
#pandas for data analysis
```

```python
r.json()['Meta Data']
```

```python
df = pd.DataFrame(r.json()['Time Series (Digital Currency Daily)']).transpose()
df
```

```python
vmdf = df.loc[:, ['4b. close (USD)','5. volume','6. market cap (USD)']]
vmdf.index = pd.to_datetime(df.index)
# vmdf.reset_index(inplace=True)
vmdf.rename(columns={'4b. close (USD)':'Closing Price($)', '5. volume':'Volume', '6. market cap (USD)':'Market Cap'}, inplace=True)

# vmdf['Day'] = pd.to_datetime(vmdf['Day'])
vmdf.index
```

```python
vmdf['Closing Price($)'].max()
```

```python
#VARIABLES WILL BE USER-SELECTED
year = 2019
month = 6
#the datetime index has a month attr and method to check if the month is in a list
#logical and with the year to get indices for month, year
monthly_info = vmdf[np.logical_and(vmdf.index.month.isin([month]), vmdf.index.year.isin([year]))] 
monthly_info
```

```python
#bokeh to plot the dataframe
from bokeh.plotting import figure, output_file, show, ColumnDataSource, save, output_notebook
from bokeh.models import Range1d
from bokeh.embed import components #for embedable code
```

```python
source = ColumnDataSource(monthly_info)
output_file('bplot.html') #THIS WILL CHANGE LATER WHEN THE PLOT IS EMBEDDED INSTEAD
btc_plot = figure(
                title = f"{payload['symbol']} Closing Price for {month}/{year}",
                x_axis_type = 'datetime',
                y_range = Range1d(0,14000)
                )

btc_plot.line(
    y = monthly_info['Closing Price($)'],
    x = monthly_info.index,


    )

show(btc_plot)
```
